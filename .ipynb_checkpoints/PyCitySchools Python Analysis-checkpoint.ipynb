```python
#Import Dependencies
import pandas as pd
```

```python
# Files to Load
PyCitySchools_Path = "Resources/schools_complete.csv"
PyCityStudents_Path = "Resources/students_complete.csv"
```

```python
# Read the modified PyCitySchools csv and store into Pandas DataFrame
PyCitySchools_df = pd.read_csv(PyCitySchools_Path)
PyCitySchools_df
```

```python
# Read the modified PyCityStudents csv and store into Pandas DataFrame
PyCityStudents_df = pd.read_csv(PyCityStudents_Path)
PyCityStudents_df
```

```python
# Combine the data into a single dataset
complete_schools_df = pd.merge(PyCityStudents_df, PyCitySchools_df, how="left", on=["school_name", "school_name"])

complete_schools_df
```

```python
#Count the total schools in PyCity
total_schools = PyCitySchools_df["School ID"].count()
total_schools
```

```python
#Count the total students in PyCity
total_students = complete_schools_df["Student ID"].count()
total_students
```

```python
#Add the total budget for all schools
total_budget = PyCitySchools_df["budget"].sum()
total_budget
```

```python
avg_math_score = round(PyCityStudents_df["math_score"].mean(),2)
avg_math_score
```

```python
avg_read_score = round(PyCityStudents_df["reading_score"].mean(),2)
avg_read_score
```

```python
#Students must have a score of 60% or higher to pass reading or math
math_pass = (PyCityStudents_df["math_score"] >= 70)

reading_pass = (PyCityStudents_df["reading_score"] >= 70)

per_math_pass = round((math_pass.sum()/total_students)*100,2)

per_reading_pass = round((reading_pass.sum()/total_students)*100,2)
per_math_pass, per_reading_pass
```

```python
complete_schools_df["Pass Math"] = (complete_schools_df["math_score"] >= 70)*100
complete_schools_df["Pass Reading"] = (complete_schools_df["reading_score"] >= 70)*100
complete_schools_df
```

```python
#Find the passing average of both math and reading

overall_pass = round(((per_math_pass+per_reading_pass)/2))
overall_pass
```

```python
#Create a district summary
district_summary = {"Total Schools" : total_schools,
                   "Total Students" : total_students,
                   "Total Budget" : total_budget,
                   "Average Math Score" : avg_math_score,
                   "Average Reading Score" : avg_read_score,
                    "% Passing Math" : per_math_pass,
                   "% Passing Reading" : per_reading_pass,
                   "% Overall Passing" : overall_pass
                  }


district_summary_df = pd.DataFrame([district_summary])                                              
                                   
district_summary_df = district_summary_df[["Total Schools",
                   "Total Students","Total Budget","Average Math Score", "Average Reading Score",
                   "% Passing Math", "% Passing Reading", "% Overall Passing" ]]

#district_summary_df["Total Students"] = district_summary_df["Total Students"].map("${:,.2f}".format)
#district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df
```

```python
district_summary_df.columns
```

```python
#Create a table with average scores for math and reading per school

avg_math_read = complete_schools_df.groupby(['school_name'])["budget","size",'reading_score', 'math_score','Pass Math','Pass Reading'].mean().reset_index()
avg_math_read

```

```python
#Find the budget per student and create a new column
avg_math_read["% Overall Passing Rate"] = (avg_math_read["Pass Math"] + avg_math_read["Pass Reading"])/2
avg_math_read["Budget per Student"] = avg_math_read['budget']/avg_math_read['size']

avg_math_read
```

```python
schools_df = pd.merge(avg_math_read, PyCitySchools_df, how="left", on=["school_name", "school_name"])
schools_df
```

```python
#Find the overall passing per school
schools_df = schools_df.rename(columns={'budget_y':"Total School Budget",'school_name':'School Name','size_y':'Total Students',"reading_score":"Avg. Reading Score", "math_score":"Avg. Math Score","Pass Math": "% Pass Math","Pass Reading":"% Pass Reading","type":"School Type"})
schools_df
```

```python
del schools_df['size_x']
del schools_df['School ID']
del schools_df['budget_x']
```

```python
schools_df
```

```python
schools_df = schools_df[["School Name",
                   "School Type","Total Students","Total School Budget","Budget per Student","Avg. Math Score", "Avg. Reading Score",
                   "% Pass Math", "% Pass Reading", "% Overall Passing Rate"]]

schools_df
```

```python
top_performing_schools_df = schools_df.sort_values(by=['% Overall Passing Rate'], ascending=False).head(5)
#top_performing_schools_df["Total School Budget"] = top_performing_schools_df["Total School Budget"].map("${:,.2f}".format)
#top_performing_schools_df["Budget per Student"] = top_performing_schools_df["Budget per Student"].map("${:,.2f}".format)

top_performing_schools_df
```

```python
bottom_performing_schools_df = schools_df.sort_values(by=['% Overall Passing Rate'], ascending=True).head(5)
#bottom_performing_schools_df["Total School Budget"] = bottom_performing_schools_df["Total School Budget"].map("${:,.2f}".format)
#bottom_performing_schools_df["Budget per Student"] = bottom_performing_schools_df["Budget per Student"].map("${:,.2f}".format)

bottom_performing_schools_df
```

```python
math_scores_by_grade_df = pd.pivot_table(PyCityStudents_df, values=['math_score'], index=['school_name'], 
                                         columns=['grade'])

math_scores_by_grade_df
```

```python
reading_scores_by_grade_df = pd.pivot_table(PyCityStudents_df, values=['reading_score'], index=['school_name'], 
                                         columns=['grade'])
reading_scores_by_grade_df
```

```python
#Create Bins
bins = [0, 585, 615, 645, 675]
spending_labels = ['$0-585', '$586-615', '$616-645', '$646-675']
```

```python
bins_school_spending = pd.cut(schools_df["Budget per Student"], bins, labels=spending_labels)

bins_school_spending = pd.DataFrame(bins_school_spending)

schools_df["Spending Level"] = bins_school_spending

bins_school_spending
```

```python
scores_by_school_spending = schools_df.groupby(['Spending Level'])['Avg. Reading Score',
                                                         'Avg. Math Score',
                                                         '% Pass Reading',
                                                         '% Pass Math',
                                                         '% Overall Passing Rate'
                                                         ].mean()
scores_by_school_spending
```

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
 
bins_school_size = pd.cut(schools_df['Total Students'], size_bins, labels=group_names)

bins_school_size = pd.DataFrame(bins_school_size)

schools_df['Total Students'] = bins_school_size

bins_school_size
```

```python
scores_by_school_size = schools_df.groupby(['Total Students'])['Avg. Reading Score',
                                                         'Avg. Math Score',
                                                         '% Pass Reading',
                                                         '% Pass Math',
                                                         '% Overall Passing Rate'
                                                         ].mean()
scores_by_school_size
```

```python
scores_by_school_type = schools_df.groupby(['School Type'])['Avg. Reading Score',
                                                         'Avg. Math Score',
                                                         '% Pass Reading',
                                                         '% Pass Math',
                                                         '% Overall Passing Rate'
                                                         ].mean()
scores_by_school_type
```
